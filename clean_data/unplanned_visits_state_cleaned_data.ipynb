```python
import pandas as pd
from pathlib import Path
Unpalnned_data_to_load = Path("../Resources/Unplanned_Hospital_Visits-State.csv")
unplanned_visits_data = pd.read_csv(Unpalnned_data_to_load)
unplanned_visits_data.head(20)
```

```python
unplanned_visits_data.columns
```

```python
clean_unplanned_visits_data = unplanned_visits_data[["State", "Measure ID", "Number of Hospitals Worse", "Number of Hospitals Same",
                                               "Number of Hospitals Better", "Number of Hospitals Too Few"]]
clean_unplanned_visits_data
```

```python
clean_unplanned_visits_data.head(20)
```

```python
measure_id = ['READM_30_AMI',
             'READM_30_CABG',
             'READM_30_COPD',
             'READM_30_HF',
             'READM_30_HIP_KNEE',
             'READM_30_HOSP_WIDE',
             'READM_30_PN']

mask = clean_unplanned_visits_data['Measure ID'].isin(measure_id)
readmission_df = clean_unplanned_visits_data[mask]
readmission_df.dtypes
```

```python

hospitals_same = pd.to_numeric(readmission_df['Number of Hospitals Same'],errors ='coerce')
hospitals_worse = pd.to_numeric(readmission_df['Number of Hospitals Worse'],errors ='coerce')
hospitals_better = pd.to_numeric(readmission_df['Number of Hospitals Better'],errors ='coerce')
hospitals_too_few = pd.to_numeric(readmission_df['Number of Hospitals Too Few'],errors ='coerce')
total_hospitals = hospitals_same + hospitals_worse + hospitals_better + hospitals_too_few



state = readmission_df["State"]

```

```python
# readmission_df = pd.to_numeric(readamission_df['Number of Hospitals Same','Number of Hospitals Same', 'Number of Hospitals Better',
#        'Number of Hospitals Too Few'],errors ='coerce')
readmission_df = pd.DataFrame({"State":state,
                               "Number of Hospitals Worse":hospitals_worse,
                               "Number of Hospitals Same": hospitals_same,
                               "Number of Hospitals Better":hospitals_better,
                               "Number of Hospitals Too Few":hospitals_too_few,
                               "Total Hospitals": total_hospitals
                                })
readmission_df.dtypes
```

```python
readmission_state = readmission_df.groupby(['State']).sum(numeric_only=True)
readmission_state
```

```python
hospitals_better = readmission_state['Number of Hospitals Better']
hospitals_worse = readmission_state['Number of Hospitals Worse']
hospitals_same = readmission_state['Number of Hospitals Same']
hospitals_too_few = readmission_state['Number of Hospitals Too Few']
total_hospitals = hospitals_same + hospitals_worse + hospitals_better + hospitals_too_few
better_performing = round((hospitals_better/total_hospitals)*100,2)
worse_performing = round((hospitals_worse/total_hospitals)*100,2)
same_as_national_performance = round((hospitals_same/total_hospitals)*100,2)

state_percentages = pd.DataFrame({
                                  "% Outstanding Hospitals": better_performing,
                                  "% Under Performing":worse_performing,
                                  "% Average Performance": same_as_national_performance})

readmission_perfomance_of_state = pd.merge(readmission_state,state_percentages, how='left', on='State')
readmission_perfomance_of_state.fillna(0)

 n
```
